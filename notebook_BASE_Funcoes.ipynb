### Histórico do aluno

```python
df = pd.read_csv("dataset_historico_aluno.csv")
df.head()
```

```python
def criar_historico(df):
    historico = {}

    df_ordenado = df.sort_values("periodo")  # garante ordem cronológica

    for linha in df_ordenado.itertuples():
        codigo = linha.codigo_disciplina
        situacao = linha.situacao

        if situacao == "APR":
            historico[codigo] = "CONCLUIDA"
        elif situacao in ["REP", "REPMF"]:
            historico[codigo] = "REPROVADA"

    return historico
```

```python
# Gera o histórico do aluno
historico_aluno = criar_historico(df)


historico_aluno
```

### Encontrar Disciplinas disponíveis

```python
def encontrar_disponiveis():

    disponiveis = []


    for disciplina in disciplinas.values():


        if disciplina.estado not in ["NAO_CURSADA", "REPROVADA"]:  

            continue


        liberada = True


        for requisito in disciplina.pre_requisitos:


            if requisito.estado != "CONCLUIDA":

                liberada = False



        if liberada:

            disciplina.estado = "DISPONIVEL"

            disponiveis.append(disciplina)



    return disponiveis
```

### Ordenação Topológica

```python
ordem = nx.topological_sort(grafo)


try:
    ordem = list(nx.topological_sort(grafo.copy()))
except nx.NetworkXUnfeasible:
    print("Aviso: o grafo possui ciclo. Usando ordem por componentes fortemente conexas.")
    ordem = []
    g_comp = nx.condensation(grafo.copy())

    for comp in nx.topological_sort(g_comp):
        membros = g_comp.nodes[comp]["members"]
        ordem.extend(sorted(membros, key=lambda d: d.codigo))

for disciplina in ordem:
    print(f"{disciplina.codigo} - {disciplina.nome}")
```

### Mostrar Grafo

```python
plt.figure(figsize=(20, 10))

# ── Mapeamento semestre → lista de códigos ──────────────────────────────────
semestres = {
    1: ["MAT00A", "XMAC01", "XDES01", "CRSC03", "CAHC04"],
    2: ["MAT00B", "CMAC04", "CTCO01", "CRSC04"],
    3: ["CTCO02", "CMAC03", "XMAC02", "XDES02", "XDES04", "CRSC02"],
    4: ["XRSC01", "CRSC05", "CTCO04", "XDES03", "CDES05", "CMAC05"],
    5: ["CTCO05", "XPAD01", "CMCO05", "XMCO01", "CTCO03"],
    6: ["XAHC02", "CTCO06", "XAHC01"],
    7: ["XAHC03", "TCC1"],
    8: ["TCC2"],
}

# ── Calcula posições: x = semestre, y distribuído verticalmente ─────────────
pos = {}
for sem, codigos in semestres.items():
    n = len(codigos)
    for i, codigo in enumerate(codigos):
        node = disciplinas.get(codigo)
        if node:
            x = sem - 1
            y = (n - 1) / 2 - i
            pos[node] = (x, y)

# ── Nós sem semestre definido ficam abaixo do grafo ─────────────────────────
sem_max = max(semestres.keys())
extra = [n for n in grafo.nodes if n not in pos]
for i, node in enumerate(extra):
    pos[node] = (sem_max + 1, -i)

# ── Cores por estado ─────────────────────────────────────────────────────────
cor_estado = {
    "CONCLUIDA":   "#4CAF50",
    "REPROVADA":   "#F44336",
    "DISPONIVEL":  "#2196F3",
    "NAO_CURSADA": "#B0BEC5",
}
node_colors = [cor_estado.get(getattr(n, "estado", "NAO_CURSADA"), "#B0BEC5")
               for n in grafo.nodes]

labels = {d: f"{d.codigo}\n{d.nome}" for d in grafo.nodes}

nx.draw(
    grafo,
    pos,
    labels=labels,
    with_labels=True,
    node_size=4000,
    node_color=node_colors,
    font_size=7.5,
    font_weight="bold",
    arrows=True,
    arrowsize=15,
    edge_color="#607D8B",
    width=1.5,
)

# ── Rótulos dos semestres no topo ────────────────────────────────────────────
ax = plt.gca()
for sem in semestres:
    ax.text(sem - 1, max(len(v) for v in semestres.values()) / 2 + 0.7,
            f"{sem}º Período",
            ha="center", va="bottom", fontsize=15, fontweight="bold", color="#37474F")

# ── Legenda ──────────────────────────────────────────────────────────────────
from matplotlib.patches import Patch
legenda = [Patch(facecolor=c, label=s) for s, c in cor_estado.items()]
plt.legend(handles=legenda, loc="lower right", fontsize=18)

plt.title("Grade Curricular – Pré-requisitos por Período", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
```

#### Análise do histórico do aluno

```python
# ── Análise do histórico: familiaridade por área e carga segura ──────────────
#
# Requer que o df já esteja carregado (cell 8) e que area_da_disciplina
# esteja definida (célula anterior).
#
# Saídas:
#   familiaridade_area : dict {area: taxa_aprovacao}  → 0.0 a 1.0
#   carga_segura_n     : int  → nº médio de disciplinas em semestres sem reprovação
#   carga_segura_horas : int  → carga horária média estimada nesses semestres

def analisar_historico(df):
    """
    Analisa o histórico do aluno para extrair:
      - familiaridade_area: proporção de aprovações por área
      - carga_segura_n    : média de disciplinas cursadas em semestres 100% aprovados
      - carga_segura_horas: estimativa de carga horária segura usando carga_horaria real
    """
    df = df.copy()
    df["area"] = df["codigo_disciplina"].apply(area_da_disciplina)

    # ── 1. Familiaridade por área ────────────────────────────────────────────
    # Usa nota_final para ponderar: aprovações com nota alta valem mais
    familiaridade = {}
    for area, grupo in df.groupby("area"):
        aprovadas = grupo[grupo["situacao"] == "APR"]
        total     = len(grupo)

        if total == 0:
            familiaridade[area] = 0.5
            continue

        # taxa base de aprovação (0..1)
        taxa_aprovacao = len(aprovadas) / total

        # bonus de desempenho: média das notas dos aprovados normalizada (0..1)
        # usa nota_final; ignora NaN (ex: trancamentos)
        notas_validas = aprovadas["nota_final"].dropna()
        if len(notas_validas) > 0:
            media_nota = notas_validas.mean() / 10.0   # normaliza para 0..1
        else:
            media_nota = 0.5

        # familiaridade = 70% peso da taxa de aprovação + 30% peso da nota média
        familiaridade[area] = round(0.7 * taxa_aprovacao + 0.3 * media_nota, 4)

    # Áreas não vistas no histórico ficam com taxa neutra (0.5)
    for area in AREAS_POR_PREFIXO.values():
        if area not in familiaridade:
            familiaridade[area] = 0.5

    # ── 2. Carga segura ──────────────────────────────────────────────────────
    # Para cada período: soma carga_horaria real das disciplinas cursadas
    # e verifica se houve reprovação
    periodos_info = []
    for periodo, grupo in df.groupby("periodo"):
        # ignora trancamentos no cômputo da carga
        grupo_ativo  = grupo[grupo["situacao"] != "TRANC"]
        teve_rep     = grupo_ativo["situacao"].isin(["REP", "REPMF"]).any()
        carga_total  = grupo_ativo["carga_horaria"].sum()
        n_disc       = len(grupo_ativo)
        periodos_info.append({
            "periodo"   : periodo,
            "n_disc"    : n_disc,
            "carga"     : carga_total,
            "teve_rep"  : teve_rep
        })

    semestres_ok = [p for p in periodos_info if not p["teve_rep"]]

    if semestres_ok:
        carga_segura_n     = round(
            sum(p["n_disc"] for p in semestres_ok) / len(semestres_ok), 1
        )
        carga_segura_horas = int(
            sum(p["carga"] for p in semestres_ok) / len(semestres_ok)
        )
    else:
        # nenhum semestre 100% aprovado → semestre com menos carga como base
        pior = min(periodos_info, key=lambda p: p["carga"]) if periodos_info else None
        carga_segura_n     = pior["n_disc"]   if pior else 3
        carga_segura_horas = pior["carga"]    if pior else 192  # 3 × 64h

    return familiaridade, carga_segura_n, carga_segura_horas, periodos_info


familiaridade_area, carga_segura_n, carga_segura_horas, periodos_info = analisar_historico(df)

# ── Exibição ─────────────────────────────────────────────────────────────────
print("Familiaridade do aluno por área:")
print(f"  {'Área':35s} {'Score':>7}  {'Barra'}")
print("  " + "─" * 60)
for area, score in sorted(familiaridade_area.items(), key=lambda x: -x[1]):
    barra = "█" * int(score * 20)
    print(f"  {area:35s} {score:>7.2f}  {barra}")

print(f"\nHistórico por período:")
print(f"  {'Período':>8}  {'Disciplinas':>12}  {'Carga(h)':>10}  {'Reprovação'}")
print("  " + "─" * 50)
for p in sorted(periodos_info, key=lambda x: x["periodo"]):
    rep = "⚠ SIM" if p["teve_rep"] else "ok"
    print(f"  {p['periodo']:>8}  {p['n_disc']:>12}  {p['carga']:>10}  {rep}")

print(f"\nCarga segura estimada : {carga_segura_n} disciplinas/semestre")
print(f"                      ≈ {carga_segura_horas}h por semestre")
```

```python
import networkx as nx
from collections import deque

# ── Pesos do score (Ajuste esses valores para calibrar o algoritmo) ──────────
PESO_DESBLOQUEIO_DIRETO   = 3.0   # cada disciplina desbloqueada diretamente
PESO_DESBLOQUEIO_INDIRETO = 1.0   # cada disciplina desbloqueada no 2º nível ou adiante
PESO_OBRIGATORIA          = 2.0   # multiplicador para disciplinas obrigatórias
PESO_OPTATIVA             = 0.5   # multiplicador para disciplinas optativas
PESO_OFERTA_ANUAL         = 1.0   # multiplicador: ofertada todo semestre
PESO_OFERTA_SEMESTRAL     = 1.5   # multiplicador: ofertada só uma vez por ano
MULT_FAMILIARIDADE_MAX    = 1.5   # multiplicador máximo quando familiaridade = 100%
PESO_CAMINHO_CRITICO      = 2.5   # NOVO: bônus massivo para matérias do caminho crítico


# ── Funções de Análise de Grafos e Valoração ─────────────────────────────────

def calcular_desbloqueios(disciplina, todas_disciplinas):
    """
    Retorna (n_diretos, n_indiretos) utilizando Busca em Largura (BFS).
    Explora toda a árvore de dependências futuras dessa disciplina.
    """
    # Constrói lista de adjacência reversa (quem depende de quem) para a BFS
    grafo_adj = {d: [] for d in todas_disciplinas}
    for d in todas_disciplinas:
        for pre_req in getattr(d, 'pre_requisitos', []):
            if pre_req in grafo_adj:
                grafo_adj[pre_req].append(d)
                
    # Desbloqueios Diretos (Arestas imediatas)
    diretos = grafo_adj.get(disciplina, [])
    n_diretos = len(diretos)
    
    # Configuração da BFS para os desbloqueios indiretos
    visitados = set(diretos)
    visitados.add(disciplina)
    
    fila = deque(diretos)
    n_indiretos = 0
    
    # Travessia BFS
    while fila:
        atual = fila.popleft()
        
        for vizinho in grafo_adj.get(atual, []):
            if vizinho not in visitados:
                visitados.add(vizinho)
                fila.append(vizinho)
                n_indiretos += 1
                
    return n_diretos, n_indiretos


def calcular_score(disciplina, todas_disciplinas, familiaridade_area, caminho_critico):
    """
    Calcula o score estratégico aplicando as heurísticas do grafo.
    """
    todas = list(todas_disciplinas.values())

    # ── 1: Alcançabilidade (Centralidade de Grau e Fecho Transitivo) ─────────
    n_diretos, n_indiretos = calcular_desbloqueios(disciplina, todas)
    score_desbloqueio = (n_diretos   * PESO_DESBLOQUEIO_DIRETO +
                         n_indiretos * PESO_DESBLOQUEIO_INDIRETO)

    # ── 2: Tipo (obrigatória / optativa) ─────────────────────────────────────
    tipo    = getattr(disciplina, "tipo", "OBRIGATORIA")
    p_tipo  = PESO_OBRIGATORIA if tipo == "OBRIGATORIA" else PESO_OPTATIVA

    # ── 3: Periodicidade de oferta ───────────────────────────────────────────
    oferta  = getattr(disciplina, "oferta", "SEMESTRAL")
    p_oferta = PESO_OFERTA_ANUAL if oferta == "ANUAL" else PESO_OFERTA_SEMESTRAL

    # ── 4: Familiaridade do aluno com a área ─────────────────────────────────
    # Nota: assume-se que area_da_disciplina(codigo) já existe no seu escopo global
    try:
        area = getattr(disciplina, "area", area_da_disciplina(disciplina.codigo))
    except NameError:
        area = getattr(disciplina, "area", "Geral")
        
    taxa_f  = familiaridade_area.get(area, 0.5)
    mult_f  = 1.0 + taxa_f * (MULT_FAMILIARIDADE_MAX - 1.0)

    # ── 5: Caminho Crítico (CPM) ─────────────────────────────────────────────
    # Aplica o bônus se a disciplina for um gargalo logístico para a formatura
    mult_critico = PESO_CAMINHO_CRITICO if disciplina.codigo in caminho_critico else 1.0

    # ── Score final ──────────────────────────────────────────────────────────
    score_final = score_desbloqueio * p_tipo * p_oferta * mult_f * mult_critico

    detalhes = {
        "diretos"          : n_diretos,
        "indiretos"        : n_indiretos,
        "score_desbloqueio": round(score_desbloqueio, 3),
        "tipo"             : tipo,
        "peso_tipo"        : p_tipo,
        "oferta"           : oferta,
        "peso_oferta"      : p_oferta,
        "area"             : area,
        "familiaridade"    : round(taxa_f, 4),
        "mult_familiaridade": round(mult_f, 4),
        "caminho_critico"  : "SIM" if disciplina.codigo in caminho_critico else "NÃO",
        "mult_critico"     : mult_critico,
        "score_final"      : round(score_final, 3),
    }

    # Atribui o score à propriedade da classe para ser lido no momento de imprimir
    disciplina.score = round(score_final, 3)

    return round(score_final, 3), detalhes


def calcular_scores_disponiveis(disponiveis, todas_disciplinas, familiaridade_area, grafo):
    """
    Função envelopadora que extrai o caminho crítico e calcula o score de todas.
    """
    # 1. Extração do Caminho Crítico do Grafo Acíclico Direcionado
    try:
        # Pega a lista de objetos Disciplina que formam o caminho crítico
        caminho_critico_obj = nx.dag_longest_path(grafo)
        
        # [CORREÇÃO AQUI]: Extrai apenas o código (string) de cada objeto
        caminho_critico = [d.codigo for d in caminho_critico_obj]
        
        # Agora o .join vai funcionar perfeitamente, pois é uma lista de textos
        print("\n[!] Caminho Crítico Calculado:", " -> ".join(caminho_critico))
    except nx.NetworkXUnfeasible:
        print("\n[ERRO] O grafo possui dependências cíclicas intransponíveis!")
        caminho_critico = []

    # 2. Avaliação de cada disciplina disponível
    ranking = []
    for d in disponiveis:
        # Passamos a lista de strings (caminho_critico) para o calculo
        score, detalhes = calcular_score(d, todas_disciplinas, familiaridade_area, caminho_critico)
        ranking.append((d, detalhes))
        
    # Ordena pelo score (Do maior para o menor peso)
    ranking.sort(key=lambda x: x[0].score, reverse=True)
    
    return ranking
```
